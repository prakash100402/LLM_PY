```python
import torch
import torch.nn as nn
from torch.nn import functional as F

torch.manual_seed(1337) 

device = 'cuda' if torch.cuda.is_available() else 'cpu'
print(device)

block_size = 8
batch_size = 32          
max_iters = 5000        
learning_rate = 1e-2     
eval_iters = 250
eval_interval = 500    
```

```python
with open('ROMEO_JULIET.txt', 'r', encoding='utf-8') as f:
    text = f.read()

chars = sorted(set(text))
print(chars)
vocab_size = len(chars)
```

```python
string_to_int = { ch:i for i,ch in enumerate(chars) }
int_to_string = { i:ch for i,ch in enumerate(chars) }
encode = lambda s: [string_to_int[c] for c in s]
decode = lambda l: ''.join([int_to_string[i] for i in l])

data = torch.tensor(encode(text), dtype=torch.long)
#print(data)
```

```python
n = int(0.8*len(data))
train_data = data[:n]
val_data = data[n:]
#print(val_data)

def get_batch(split):
    data = train_data if split == 'train' else val_data
    ix = torch.randint(len(data) - block_size, (batch_size,))
    x = torch.stack([data[i:i+block_size] for i in ix])
    y = torch.stack([data[i+1:i+block_size+1] for i in ix])
    x, y = x.to(device), y.to(device)
    return x, y

x, y = get_batch('train')
print('inputs:')
# print(x.shape)
print(x)
print('targets:')
print(y)
```

```python
@torch.no_grad()
def estimate_loss():
    out = {}
    model.eval()
    for split in ['train', 'val']:
        losses = torch.zeros(eval_iters)
        for k in range(eval_iters):
            X, Y = get_batch(split)
            logits, loss = model(X, Y)
            losses[k] = loss.item()
        out[split] = losses.mean()
    model.train()
    return out
```

```python
class BigramLanguageModel(nn.Module):
    def __init__(self, vocab_size):
        super().__init__()
        self.token_embedding_table = nn.Embedding(vocab_size, vocab_size)
 
    def forward(self, index, targets=None):
        logits = self.token_embedding_table(index)
 
        if targets is None:
            loss = None
        else:
            B, T, C = logits.shape
            logits = logits.view(B * T, C)
            targets = targets.view(B * T)
            loss = F.cross_entropy(logits, targets)
 
        return logits, loss
 
    def generate(self, index, max_new_tokens):
        # index is (B, T) array of indices in the current context
        for _ in range(max_new_tokens):
            logits, loss = self.forward(index)
            logits = logits[:, -1, :]                    # (B, C)
            probs = F.softmax(logits, dim=-1)             # (B, C)
            index_next = torch.multinomial(probs, num_samples=1)  # (B, 1)
            index = torch.cat((index, index_next), dim=1)  # (B, T+1)
        
        return index

model = BigramLanguageModel(vocab_size)
m = model.to(device)

context = torch.zeros((1,1), dtype=torch.long, device=device)
generated_chars = decode(m.generate(context, max_new_tokens=500)[0].tolist())
print(generated_chars)
```

```python

# Train two versions and keep only the better one
# Trains an "old settings" model (V1) and a "new/optimized settings" model (V2),
# compares their val loss, and saves only the winner as bigram_model_best.pt
 
# --- V1: old settings ---
model_v1 = BigramLanguageModel(vocab_size)
m = model_v1.to(device)
optimizer = torch.optim.AdamW(model_v1.parameters(), lr=3e-4)
 
for iter in range(1000):
    xb, yb = get_batch('train')
    model = model_v1  # estimate_loss() reads the global `model` variable
    logits, loss = model_v1.forward(xb, yb)
    optimizer.zero_grad(set_to_none=True)
    loss.backward()
    optimizer.step()
 
model = model_v1
v1_losses = estimate_loss()
print("V1 val loss:", v1_losses['val'].item())
 
# --- V2: optimized settings ---
model_v2 = BigramLanguageModel(vocab_size)
m = model_v2.to(device)
optimizer = torch.optim.AdamW(model_v2.parameters(), lr=1e-2)
 
for iter in range(5000):
    xb, yb = get_batch('train')
    model = model_v2
    logits, loss = model_v2.forward(xb, yb)
    optimizer.zero_grad(set_to_none=True)
    loss.backward()
    optimizer.step()
 
model = model_v2
v2_losses = estimate_loss()
print("V2 val loss:", v2_losses['val'].item())
 
# --- Compare and save only the winner ---
if v2_losses['val'] < v1_losses['val']:
    best_model = model_v2
    print(f"V2 wins ({v2_losses['val']:.3f} < {v1_losses['val']:.3f}) -- saving V2")
else:
    best_model = model_v1
    print(f"V1 wins ({v1_losses['val']:.3f} <= {v2_losses['val']:.3f}) -- saving V1")
 
torch.save(best_model.state_dict(), 'bigram_model_best.pt')
 
# point working variables at the winner
model = best_model
m = model.to(device)
```

```python
current_losses = estimate_loss()  # loss of the model currently in `model`/`m`
current_val = current_losses['val'].item()
 
temp_model = BigramLanguageModel(vocab_size)
temp_model.load_state_dict(torch.load('bigram_model.pt'))
temp_model = temp_model.to(device)
 
model_backup = model
model = temp_model
saved_losses = estimate_loss()
saved_val = saved_losses['val'].item()
model = model_backup  # restore in-memory model as the active one
 
print(f"In-memory val loss: {current_val:.3f}")
print(f"Saved checkpoint val loss: {saved_val:.3f}")
 
if current_val < saved_val:
    torch.save(model.state_dict(), 'bigram_model.pt')
    print("In-memory model is better -- overwrote checkpoint.")
else:
    print("Saved checkpoint is already better -- nothing changed.")
```

```python
optimizer = torch.optim.AdamW(model.parameters(), lr=learning_rate)

for iter in range(max_iters):
    if iter % eval_interval == 0 or iter == max_iters - 1:
        losses = estimate_loss()
        print(f"step: {iter}, train loss: {losses['train']:.3f}, val loss: {losses['val']:.3f}")

    xb, yb = get_batch('train')

    logits, loss = model.forward(xb, yb)
    optimizer.zero_grad(set_to_none=True)
    loss.backward()
    optimizer.step()

print(loss.item())

# save the trained model so you don't lose it when the kernel restarts
torch.save(model.state_dict(), 'bigram_model.pt')
print("model saved to bigram_model.pt")int(loss.item())
```
